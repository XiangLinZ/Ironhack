```python
#Importo todo lo que necesito
from pygbif import registry
from pygbif import species
from pygbif import occurrences
import pandas as pd
```

```python
#Mi API es una librer√≠a de llamadas ilimitadas
#pd_info significa informaci√≥n del panda, he usado la llamada "species" con la subllamada "name_backbone"
#Esto, me da todas las especies que hay y backbone es para indicarle un nombre aprox
#En mi caso es el panda, que me gustan üêº, su nombre cient√≠fico lo he buscado en internet
#name_backbone, si tu nombre no es del todo correcto te lo corrige, o busca en su base de datos un nombre con
#altas coincidencias
# deberr√≠a de haber usado triple comillas... lo s√©...
pd_info = species.name_backbone(name = "Ailuropoda melanoleuca")
pd_info
```

```python
""" Lo de antes lo he hecho porque realmente solo quiero la usageKey del panda
    es decir, necesito su c√≥digo por el que lo guarda, por as√≠ decirlo,
    este c√≥digo me servir√° luego para otra llamada que quer√≠a hacer en un principio
"""
keyPD = pd_info["usageKey"]
keyPD
```

```python
""" Con la  key de antes, del panda, ahora hago otra llamada, "ocurrences" esta llamada te devuelve las veces
    que se ha avistado un panda con toda la info del avistamiento, pero yo no quiero una sola ocurrencia,
    por lo que uso el m√©todo search para que me devuelva todos los avistamientos, para no pasarme de loco
    le he dicho que el m√°ximo de ocurrencias que me devuelva sean 100
    y como taxonKey, le doy la key del panda, que he buscado en la llamada anterior
"""
res = occurrences.search(taxonKey = keyPD, limit = 100)
```

```python
""" La respuesta es bastante loca, tiene en realidad mucha m√°s info de la que he comentado antes
    pero solo me interesa los resultados, es decir, las ocurrencias como tal, no toda la info de alrededor
    por lo que del diccionario que me devuelve, me quedo solo con "results" y lo almaceno (me lo da en formato lista)
"""
lista_ocurrencias = res["results"]
```

```python
""" transformo mi lista de ocurrencias en un dataframe
    y muestro todas las columnas para explorar y decidir con qu√© columnas quedarme
    he hecho otras cosas, pero las he borrado, tipo contar los nulos de las columnas
    y los types, porque... madre de dios, que cantidad de info random que no necesito... 
    T_T
"""

df0 = pd.DataFrame(lista_ocurrencias)
pd.options.display.max_columns = None
df0.head()
```

```python
""" Aqu√≠, creo un DF nuevo con solo las columnas que me interesan...
    bueno... he visto que no est√° lleno de nulos o de info que es un poco menos √∫til
    y lo meto en forma de lista de nuevas columnas en el nuevo df
"""
df1 = df0[["key","publishingCountry","lastCrawled","crawlId","basisOfRecord","taxonRank","eventDate","country","identifier"]]
df1.head()
```
