**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# 1. your code here
def uniform_rng(botton, ceiling, count):
    """A function that return a sample with a uniform distribution

    Args:
        botton (Float): The min value of the sample
        ceiling (Float): The Max value of the sample
        count (Int): The size of the sample

    Returns:
        Array: A random sample with a uniform distribution in a array form
    """
    x = uniform.rvs(size = count)
    a = botton
    b = ceiling
    return a + (b-a)*x
```

```python
# 2. your code here
uniform_array0 = uniform_rng(10, 15, 100)
uniform_array1 = uniform_rng(10, 60, 1000)
```

```python
# 3. your code here
fig, ax = plt.subplots(1, 2, sharey="row")
arrays = [uniform_array0, uniform_array1]
for i in range(0,2):
    sns.histplot(arrays[i], kde = True, color= "b", bins = 10, ax=ax[i])
plt.tight_layout()
plt.ylabel("");

# It's not the same, but is close to the example
```

How are the two distributions different?

```python
## your answer below
# Well, the range it's not the same, also the sample size it's not the same, but both follow a uniform distribution
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
# 1. your code here
def normal_rng(avg, std, count):
    """A function that return a sample with a normal distribution

    Args:
        avg (Float): The avg value of our data
        std (Float): The std value of our data
        count (Int): The size of the sample

    Returns:
        Array: A random sample with a normal distribution in a array form
    """
    my_normal = norm(loc = avg, scale = std)
    return my_normal.rvs(size = count)
```

```python
# 2. your code here
normal1 = normal_rng(10, 1, 1000)
```

```python
# 3. your code here
normal2 = normal_rng(10, 50, 1000)
```

```python
# 4. your code here
fig, ax = plt.subplots(1, 2, sharey="row")
arrays_n = [normal1, normal2]
for i in range(0,2):
    sns.histplot(arrays_n[i], kde = True, color= "b", bins = 40, ax=ax[i])
plt.tight_layout()
plt.ylabel("");
```

How are the two distributions different?

```python
## your answer below
# Both are normal distribution, since our sample size it's the same, we have similar results, 
# but the range of values are quite different
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# 1 your code here
def exponential_rng(count, avg = 10):
    """A function that return a sample with a exponential distribution

    Args:
        count (Int): Size of the sample
        avg (int, optional): Avg value of our data. Defaults to 10.


    Returns:
        Array: A random sample with a exponential distribution in a array form
    """
    return np.random.exponential(scale = avg, size = count)
```

```python
# 2 your code here
exponential1 = exponential_rng(10)
exponential2 = exponential_rng(100)
```

```python
# 3 your code here
fig, ax = plt.subplots(1, 2, sharey="row")
arrays_ex = [exponential1, exponential2]
for i in range(0,2):
    sns.histplot(arrays_ex[i], kde = True, color= "b", ax=ax[i])
plt.tight_layout()
plt.ylabel("");
```
