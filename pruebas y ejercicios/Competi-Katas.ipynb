````python
def ejercicio2 (lista1, lista2):
    newlista = []
    for word in lista1:
        for word2 in lista2:
            if word in word2:
                newlista.append(word)
        
    return list(set(newlista))
````

# Escribid una función que reciba una lista de 10 enteros (entre 0 y 9), que devuelva un *string* de esos números en forma de número de teléfono.

Debe funcionar en:

```python
 "(123) 456-7890"
```

Ejemplo 2

````python
def ejercicio4(lista):
    inicio = ""
    medio = ""
    final = ""
    for num in lista[:3]:
        inicio += str(num)

    for num in lista[3:6]:
        medio += str(num)
    
    for num in lista[-4:]:
        final += str(num)
    
    return f"({inicio}) {medio}-{final}"
````

````python
#Sin clase
def ejercicio5 (lista1, lista2):
    lista1_uniq = list(set(lista1))
    for numb in lista2:
        lista1_uniq.remove(numb)
    lista_orden_normal = [numb for numb in lista1 if numb in lista1_uniq]
    
    return lista_orden_normal

#Con clase
class Resta:
    def __init__(self, lista1, lista2):
        self.lista1 = lista1
        self.lista2 = lista2
    def resta(self):
        return [numb for numb in self.lista1 if numb not in self.lista2]
````
